# If-Else 判斷式說明

`if-else` 判斷式用於根據條件執行不同的代碼區塊。
當條件為 `True` 時，執行 `if` 區塊；當條件為 `False` 時，執行 `else` 區塊（如果存在）。

### 語法:
```python
if 條件:
    # 當條件為 True 時執行的代碼
else:
    # 當條件為 False 時執行的代碼
```

### 範例 1: 基本的 If-Else 判斷式

在下面的範例中，程式會檢查一個數字是否為正數。


````python
# 範例 1: 基本的 If-Else 判斷式
number = 10

if number > 0:
    print(f"{number} 是正數。")
else:
    print(f"{number} 不是正數。")
````

### 範例 2: If-Elif-Else 判斷式

在這個範例中，我們可以使用 `elif` 來檢查多個條件。程式會檢查一個數字是正數、負數還是零。


````python
# 範例 2: If-Elif-Else 判斷式
number = -5

if number > 0:
    print(f"{number} 是正數。")
elif number == 0:
    print("這個數字是零。")
else:
    print(f"{number} 是負數。")
````

### 說明:
- `if` 語句用於檢查條件是否為真。
- `elif`（else if）語句允許檢查多個條件。
- 如果 `if` 和 `elif` 條件都不成立，則執行 `else` 區塊。

這種結構可根據不同的條件來控制程式的流程。
