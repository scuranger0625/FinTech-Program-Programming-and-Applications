# 控制結構（Control Structures）簡介

控制結構是用於控制程式執行流程的方式，根據條件或需求，決定代碼的執行順序。
控制結構分為三大類型：順序結構、選擇結構和迴圈結構。


## 1. 順序結構（Sequential Control Structure）
程式碼按書寫順序，從上到下逐行執行。


```python
# 順序結構範例
print("這是第一行")
print("這是第二行")
```

## 2. 選擇結構（Selection Control Structure）
選擇結構允許根據條件選擇性執行代碼。最常見的形式是 `if-else`。


```python
# 選擇結構範例
number = 10

if number > 0:
    print("這是正數")
else:
    print("這不是正數")
```

## 3. 迴圈結構（Repetition Control Structure）
迴圈結構允許重複執行代碼，直到條件不再滿足。常見的迴圈包括 `for` 和 `while`。


```python
# 迴圈結構範例 - for 迴圈
for i in range(5):
    print(f"第 {i+1} 次迴圈")
```

```python
# 迴圈結構範例 - while 迴圈
i = 0
while i < 5:
    print(f"第 {i+1} 次迴圈")
    i += 1
```

## 跳轉控制
在迴圈中，有時需要中斷或跳過當前的迭代，可以使用 `break` 或 `continue`。


```python
# 使用 break 和 continue
for i in range(5):
    if i == 3:
        break  # 中斷迴圈
    if i == 1:
        continue  # 跳過當前迴圈
    print(f"第 {i+1} 次迴圈")
```
